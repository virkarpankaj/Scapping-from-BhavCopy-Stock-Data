```python
import requests
import datetime
import io
import zipfile
import pandas as pd
import glob
```

## Extracting the Files

```python
months = ["JAN" , "FEB" , "MAR" , "APR" , "JUN" , "JUL" , "AUG" , "SEP" , "OCT" , "NOV" , "DEC"]

dates = range(1,31)

for month in months:
    
    for date in dates:
        
        if date < 10:
            datestring = str(0)+str(date)
        else:
            datestring = str(date)
            
        url = "https://www1.nseindia.com/content/historical/EQUITIES/2021/"+month+"/cm"+datestring+month+"2021bhav.csv.zip" 

        headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}

        r = requests.get(url,headers = headers)
        
        temp = r.status_code
        
        if (temp != 404):
            z = zipfile.ZipFile(io.BytesIO(r.content))
            z.extractall("/Equities")
        else:
            pass    
```

## Combining as 1 FIle

```python
path = r'\Equities'

allExtractedFiles = glob.glob(path+"\*.csv")

allExtractedFiles
```

```python
# reading the files one by one

list_ = []

for file in allExtractedFiles :
    
    df = pd.read_csv(file, index_col= None, header=0, engine= 'python')
    
    list_.append(df)
    
frame = pd.concat(list_, axis=0, ignore_index= True)


# saving the files as csv - exporting as csv

frame.to_csv('MasterFile.csv', index = False)

stock_data = pd.read_csv('MasterFile.csv')           # reading the MasterFile data & storing in variable
```
